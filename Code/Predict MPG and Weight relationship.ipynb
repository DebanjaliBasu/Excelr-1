Fitting a regression model.

```python
import statsmodels.formula.api as smf
model = smf.ols("hp~mpg",data = data).fit()
```

```python
model.summary()
```

#### Predicting MPG based on Hp.

```python
ypred = model.predict(data.mpg)
```
