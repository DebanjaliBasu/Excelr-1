```python
import pandas as pd
import numpy as np
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
book = pd.read_csv('/Users/halders/Documents/Excelr/dataset/book 2.csv')
book = book.drop('Unnamed: 0',axis = 1)
book = book.rename(columns = {"User.ID":"UserId","Book.Title":"Book_Title","Book.Rating":"Book_Rating"})
```

```python
#finding out length of unique user ids\
len(book.UserId.unique())
```

```python
#lenght of unique book titles
len(book.Book_Title.unique())
```

```python
book_pivot = book.pivot_table(index='UserId',columns='Book_Title',values='Book_Rating')
book_pivot
```

```python
book_pivot= book_pivot.fillna(0)
book_pivot
```

```python
#Calculating Cosine Similarity between Users
book_user_sim = 1 - pairwise_distances(book_pivot.values,metric='cosine')
book_user_sim
```

```python
#filling daigonal values with 0 
np.fill_diagonal(book_user_sim, 0)
```

```python
#storing results in a dataframe
book_us = pd.DataFrame(book_user_sim)
book_us.index = book.UserId.unique()
book_us.columns = book.UserId.unique()
book_us.iloc[0:5,0:5]
```

```python
#getting maximam similarities in every rows of users
book_us.idxmax(axis=1)[0:50]
```

```python
#defining function to get similar books read by two users
def get_similar_books( user1,user2):
    # Inner join between movies watched between two users will give 
    # the common movies watched.
    common_books = book[book.UserId == user1].merge ( book[book.UserId==user2],on = 'Book_Title', how='inner')
    return common_books.merge(book,on = 'Book_Title')
```

```python
#geting similar books for 2 users:
get_similar_books(278202,276798)
```
