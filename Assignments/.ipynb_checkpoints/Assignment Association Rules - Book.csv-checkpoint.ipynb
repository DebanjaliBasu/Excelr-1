```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
book = pd.read_csv("/Users/halders/Documents/Excelr/dataset/book.csv")
book.head()
```

```python
book.info()
```

### as we don't have any categorical data we dont have to pre process the data, and null values are also not present


## Apriori Algorithm

```python
### changing support min. support threshhold values and checkinh results
```

```python
frequent_items = apriori(book,min_support = 0.04, use_colnames = True)
frequent_items
```

```python
frequent_items = apriori(book,min_support = 0.1, use_colnames = True)
frequent_items
```

```python
frequent_items = apriori(book, min_support = 0.005, use_colnames = True)
frequent_items
```

## Creating association rules

```python
rules = association_rules(frequent_items,min_threshold=0.5,metric='lift')
rules
```

```python
# changing confidence threshhold
## This gives us the rules with threshold greater than 1


rules = association_rules(frequent_items,min_threshold=1,metric='lift')
rules
```

```python
# sorting rules based on lift.

rules.sort_values('lift', ascending = False)

```

```python
## To eliminate the reducdancy in rules

def to_list(i):
    return sorted(i)

ma1 = rules.antecedents.apply(to_list)+rules.consequents.apply(to_list)

ma1 = ma1.apply(sorted)

return_rules = list(ma1)
unique_rules = [list(m) for m in set(tuple(i) for i in return_rules)]

index_rules = []
for i in unique_rules:
    index_rules.append(return_rules.index(i))

```

```python
##Getting the rules without any redundancies
rules_without_redundancies = rules.iloc[index_rules, : ]
```

```python
#getting rules without redundancies 
rules_without_redundancies
```

```python
#stored list of values of rules without duplicates.
rules_without_redundancies.sort_values('lift',ascending=False)
```

```python
#interactive 3d plot visualizing lift, support and confidence.
import plotly.express as px
df = px.data.iris()
fig = px.scatter_3d(rules_without_redundancies, x='support', y='confidence', z='lift',
                    color='consequents',opacity=0.7)
fig.show()
```

```python
# changing min support values to 0.007 with maxium items set to 3 this time.
frequent_items = apriori(book, min_support = 0.007,max_len=3, use_colnames = True)
frequent_items
```
