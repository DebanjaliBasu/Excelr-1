```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#creating x variable and storing all the values 
x = pd.Series([24.23,25.53,25.41,24.14,29.62,28.25,25.81,24.39,40.26,32.95,91.36,25.99,39.42,26.71,35.00])
```

```python
#creating 'name' variable and storing all company names
name = pd.Series(['Allied Signal','Bankers Trust','General Mills','ITT Industries','J.P.Morgan & Co.','Lehman Brothers',
      'Marriott','MCI','Merrill Lynch','Microsoft','Morgan Stanley','Sun Microsystems','Travelers','US Airways',
      'Warner-Lambert'])
```

```python
#finding out mean of the series
x.mean()
```

```python
#finding out standard deviation of the series
x.std()
```

```python
#finding out variance of the series
x.var()
```

```python
# creating a pie plot of the series
plt.figure(figsize=(6,8))
plt.pie(x,labels=name,autopct='%.1f') #autopct means auto percentage to show on each pie bar and .1f is the number of digits after decimal
plt.show()
```

```python
#finding out outlier
sns.boxplot(x)
```

#### hence we can say that 91.36 is an outlier
